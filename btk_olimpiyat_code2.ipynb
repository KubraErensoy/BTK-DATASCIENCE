```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter
import warnings
warnings.filterwarnings("ignore")
```

```python
veri=pd.read_csv("olimpiyatlar_temizlenmis.csv")
```

```python
veri.head(3)
```

```python
def sacilimGrafikciz():
  sns.scatterplot(x='boy',y='kilo',data=veri)
  plt.title('Boy ve Ağırlık İlişkişi-Beyaz Izgara Tema')
  plt.show()
sacilimGrafikciz() 
```

```python
sns.set_style("white")
sacilimGrafikciz() 
```

```python
sns.set_style("darkgrid")
sacilimGrafikciz() 
```

```python
sns.set_style("whitegrid")
sacilimGrafikciz() 
#bu grafikte boy ve kilo arasında pozitif ilişki olduğu söyleriz
```

```python
sns.scatterplot(x='boy',y='kilo',hue='madalya',data=veri)
plt.title('Madalyaya Göre Boy ve Ağırlık Dağılımı')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='boy',y='kilo',hue='cinsiyet',style='madalya',data=veri)
plt.xlabel('Sporcuların boyu')
plt.ylabel('Sporcuların Ağırlığı')
plt.title('Madalyalı Sporcuların Ağırlık-Boy Değişimi')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='boy',y='kilo',hue='cinsiyet',style='madalya',size='yas',sizes=(15,200),data=veri)
plt.xlabel('Sporcuların boyu')
plt.ylabel('Sporcuların Ağırlığı')
plt.title('Madalyalı Sporcuların Ağırlık-Boy Değişimi')
plt.show()
```

Saçılım Grafiği-Doğrusal Regrsyon

```python
sns.regplot(x='boy',y='kilo',data=veri,marker='+',scatter_kws={'alpha':0.2})
plt.title('Boy ve Ağırlık Dağılımı')
plt.show()
```

```python
#Renk Paletleri
sns.scatterplot(x='boy',y='kilo',hue='madalya',data=veri,palette='Set1')

plt.title('Madalyalı Sporcuların Ağırlık-Boy Dağılımı')
plt.show()

```

```python
#LinePlot

sns.lineplot(x='boy',y='kilo',data=veri)
plt.title('Boy ve Kilo')
plt.show()
```

```python
#Cizgi Grafik-Kategorik
sns.lineplot(x='boy',y='kilo',hue='cinsiyet',data=veri)
plt.title('Cinsiyete Göre Boy ve Ağırlık İlişkileri')
plt.show()
```

```python
sns.lineplot(x='boy',y='kilo',hue='sezon',data=veri)
plt.title('Sezona Göre Boy ve Ağırlık İlişkileri')
plt.show()
```

Histogram

```python
sns.displot(veri,x='kilo',hue='cinsiyet')
plt.ylabel("Frekans")
plt.title("Cinsiyete Göre Ayrıştırılmış Ağırlık Histogramı")
plt.show()
```

```python
#İki Boyutlu Histogram

sns.displot(veri,x='kilo',y="boy",kind="kde")
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Ağırlık-Boy Histogramı")
plt.show()
```

```python
#İki Boyutlu Histogram

sns.displot(veri,x='kilo',y="boy",hue="cinsiyet",kind="kde")
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Cinsiyete Göre Ayrıştırılmış Ağırlık-Boy Histogramı")
plt.show()
```

```python
#İki Boyutlu Histogram

sns.kdeplot(data=veri,x='kilo',y="boy",hue="cinsiyet",fill=True)
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Cinsiyete Göre Ayrıştırılmış Ağırlık-Boy Histogramı")
plt.show()
```

İki boyutlu histogram,kadınlar ile erkekler arasında boy-kilo ilişkisinin farklı gösterdiği görülebilr.

**ISI HARİTASI**

ısı haritası ,sayısal özellikler/sutunlar arasındaki ilişkiyi görselleştirmek için kullanılan bir tekniktir.

Sayısal özellikler olan boy,kilo yaş ve yıllar arasındaki ilişkiyi incelemek için korelasyon metric

```python
#ortaklık grafiği
sns.jointplot(data=veri,x="kilo",y="boy",hue="sezon",kind="kde")
plt.show()
```

```python
#
g=sns.JointGrid(data=veri,x="kilo",y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

**Çift/Eş Grafiği**

```python
sns.pairplot(veri)
plt.show()
```

```python
g=sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot,fill=True)
g.map_diag(sns.histplot,kde=True)
```
